```python
import numpy as np #Import relevant packages
import pandas as pd
import math
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import glob
import csv
import matplotlib.pyplot as plt

def unpack_data(filepath): #Returns a dictionary where each item is a df containing one run. One dict per generation.
    all_files = glob.glob(filepath + "/*.csv")
    name_list = []
    datadict = {}
    for f in all_files:
        name_list.append(f[77:-4])
    for i,n in enumerate(name_list):
        datadict[n] = pd.read_csv(all_files[i], header=0)
    for k, v in datadict.items():
        v.drop(columns=["Unnamed: 0", "volume", "spread", "10_MA", "50_MA"], inplace=True)
    return datadict

def format_data(datadict, window): #Turn dictionaries of data into features and samples lists
    x_data = []
    y_data = []
    for k, v in datadict.items():
        for i in range(len(v)):
            if i >= window:
                y_data.append(v.iloc[i]["trading_price"])
                xi = v.iloc[i-window:i].to_numpy()
                x_data.append(xi)
    x_data = np.array(x_data)
    y_data = np.array(y_data)
    y_data = np.reshape(y_data, (y_data.shape[0], 1))
    return x_data, y_data

def normalise_data(xarray, yarray): #Take the natural log and then normalise prices
    yarray_log = np.log(yarray)
    xarray_log = np.log(xarray)
    xsc = StandardScaler()
    instances, timesteps, features = xarray_log.shape
    xarray_log = np.reshape(xarray_log, (-1, features))
    xarray_norm = xsc.fit_transform(xarray_log)
    xarray_norm = np.reshape(xarray_norm, (instances, timesteps, features))
    ysc = StandardScaler().fit(yarray_log)
    yarray_norm = ysc.transform(yarray_log)
    return xarray_norm, yarray_norm, xsc, ysc

def split_data(xarray, yarray, trainratio): #Split the data into training and test sets
    train_len = int(len(xarray)*trainratio)
    test_len = len(xarray) - train_len
    x_train, y_train, x_test, y_test = xarray[:train_len], yarray[:train_len], xarray[train_len:], yarray[train_len:]
    return x_train, y_train, x_test, y_test
```

```python
gen26 = unpack_data("/Users/karangarg/Documents/Year 3 Modules/EC331/Code/rae_repo/simulations/gen26_sims/data")
```

```python
X27, y27 = format_data(gen26, 20)
print(X27.shape)
print(y27.shape)
```

```python
X27_train, y27_train, X27_test, y27_test = split_data(X27, y27, 0.8)
```

```python
X27_train, y27_train, X27scale, y27scale = normalise_data(X27_train, y27_train)
X27_test, y27_test, _, _ = normalise_data(X27_test, y27_test)
```

```python
print(X27scale.mean_)
print(X27scale.var_)
print(y27scale.mean_)
print(y27scale.var_)
```

```python
lstm27=keras.Sequential() #Set up the architecture of the model

lstm27.add(layers.LSTM(units=16, input_shape=(X27_train.shape[1], X27_train.shape[2])))
lstm27.add(layers.Dropout(0.2))

lstm27.add(layers.Dense(units=1))

print(lstm27.summary())
```

```python
lstm27.compile(optimizer='nadam', loss='mean_squared_error') #Compile and train the model

history27 = lstm27.fit(X27_train, y27_train, epochs = 10, batch_size = 32, validation_split=0.2, shuffle=False)
```

```python
lstm27.save("/Users/karangarg/Documents/Year 3 Modules/EC331/Code/rae_repo/lstms/lstm27")
```

```python
plt.figure(figsize=(15,6))
plt.plot(history27.history["loss"], label="train")
plt.plot(history27.history["val_loss"], label="validation")
plt.legend()
```

```python
y27_pred = lstm27.predict(X27_test)
y27_train_inv = y27scale.inverse_transform(y27_train.reshape(1, -1))
y27_test_inv = y27scale.inverse_transform(y27_test.reshape(1, -1))
y27_pred_inv = y27scale.inverse_transform(y27_pred)
```

```python
plt.figure(figsize=(15,6))
plt.plot(y27_test_inv.flatten(), label="Test data")
plt.plot(y27_pred_inv.flatten(), label="Model prediction")
plt.legend()
```
